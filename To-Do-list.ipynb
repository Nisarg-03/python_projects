```python
tasks = []

while True:
      print("------To Do List------")
      print("1.Add task")
      print("2.Veiw task")
      print("3.Delete Task")
      print("4.Exit")

      choice = input("Enter the choice")
     
      if choice == "1":
            task = input("Enter the task")
            tasks.append(task)
      if choice == "2":
            if not tasks:
                  print("No tasks Available")
            else:
                  print("Your Tasks:")
                  for index,task in enumerate(tasks,start=1):
                    print(f"{index}.{task}")
      if choice == "3":
            print("Your tasks :")
            for index,task in enumerate(tasks,start=1):
                  print(f"{index},{task}")   

            delete_task =  int(input("Enter task numer to delete :"))
            tasks.pop(delete_task - 1)
      if choice == "4":
            print("Exiting program")
            break
```
